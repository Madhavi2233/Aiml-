```
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['Cancer'] = [data.target_names[i] for i in data.target]
l = len(df.columns) - 1
for i in range(l):
    plt.figure(figsize=(8, 6))
    sns.set_style('whitegrid')
    feature = df.columns[i]
    sns.countplot(x='Cancer', hue=feature, data=df, palette='RdBu_r')
    plt.title(f'Countplot for {feature}')
    plt.legend(title=feature, loc='upper right')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```
for feature in data.feature_names:
    plt.figure(figsize=(10, 6))
    sns.set_style('whitegrid')
    sns.boxplot(x='Cancer', y=feature, data=df, palette='Set3')
    plt.title(f'Boxplot of {feature} by Cancer Type')
    plt.xlabel('Cancer Type')
    plt.ylabel(feature)
    plt.tight_layout()
    plt.show()
```
